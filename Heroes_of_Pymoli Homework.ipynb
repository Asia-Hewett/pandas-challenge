```python
import pandas as pd
import numpy as np
```

```python
file_to_load = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(file_to_load)
```

```python
#by using the ".nunique" function I was able to pull the total number of UNIQUE screen names from the data set
#to conclude, this means some people made multiple purchases (there are 780 entries)
total_players = purchase_data["SN"].nunique()
total_players
```

```python
#find the number of unique items
#find the average price of those unique items
#find the total number of purchases
#find the total amount of revenue earned
purchasing_analysis = {
    "Number of Unique Items": [purchase_data["Item Name"].nunique()],
    "Average Price": f"${round(purchase_data['Price'].mean(), 2)}",
    "Number of Purchases":  purchase_data["Item Name"].value_counts().sum(),
    "Total Revenue": f"${purchase_data['Price'].sum()}",
}
pd.DataFrame(purchasing_analysis)
```

```python
screen_name_df = purchase_data["SN"].nunique()

gender_count_df = {
    "Male": purchase_data.loc[purchase_data["Gender"] == "Male", :].nunique(),
    "Female": purchase_data.loc[purchase_data["Gender"] == "Female", :].nunique(),
    "Other / Non-Disclosed": purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :].nunique(),
}

numbers_for_genders = {
    "Male": gender_count_df["Male"]["SN"],
    "Female": gender_count_df["Female"]["SN"],
    "Other / Non-Disclosed": gender_count_df["Other / Non-Disclosed"]["SN"]
}

percentage_of_genders = {
    "Male": str((numbers_for_genders["Male"] / screen_name_df*100).round(2)) + "%", 
    "Female": (numbers_for_genders["Female"] / screen_name_df*100).round(2),
    "Other / Non-Disclosed": (numbers_for_genders["Other / Non-Disclosed"] / screen_name_df*100).round(2)
}

pd.DataFrame({"Total Count": numbers_for_genders, "Percentage of Players": percentage_of_genders})
```

```python
purchase_count_by_gender = {
    "Males": purchase_data.loc[purchase_data["Gender"] == "Male", :].count(),
    "Females": purchase_data.loc[purchase_data["Gender"] == "Female", :].count(),
    "Other / Non-Disclosed": purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :].count(),
}

final_purchase_count_by_gender = {
    "Males": purchase_count_by_gender["Males"]["Purchase ID"],
    "Females": purchase_count_by_gender["Females"]["Purchase ID"],
    "Other / Non-Disclosed": purchase_count_by_gender["Other / Non-Disclosed"]["Purchase ID"],
}

average_purchase_price = {
    "Males": ((purchase_data.loc[purchase_data["Gender"] == "Male", :]["Price"].sum())/(purchase_count_by_gender["Males"]["Purchase ID"])).round(2),
    "Females": ((purchase_data.loc[purchase_data["Gender"] == "Female", :]["Price"].sum())/(purchase_count_by_gender["Females"]["Purchase ID"])).round(2),
    "Other / Non-Disclosed": ((purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :]["Price"].sum())/(purchase_count_by_gender["Other / Non-Disclosed"]["Purchase ID"])).round(2)
}

total_purchase_value = {
    "Males": (purchase_data.loc[purchase_data["Gender"] == "Male", :]["Price"].sum()),
    "Females": (purchase_data.loc[purchase_data["Gender"] == "Female", :]["Price"].sum()),
    "Other / Non-Disclosed": (purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :]["Price"].sum())
}

avg_total_purchase_per_person = {
    "Males":  (total_purchase_value["Males"] / numbers_for_genders["Male"]).round(2),
    "Females": (total_purchase_value["Females"] / numbers_for_genders["Female"]).round(2),
    "Other / Non-Disclosed": (total_purchase_value["Other / Non-Disclosed"] / numbers_for_genders["Other / Non-Disclosed"]).round(2)
}

purchasing_analysis_gender = {
    "Purchase Count": final_purchase_count_by_gender,
    "Average Purchase Price": average_purchase_price,
    "Total Purchase Value": total_purchase_value,
    "Avg Total Purchase per Person": avg_total_purchase_per_person
}

pd.DataFrame(purchasing_analysis_gender)
```

```python
# Establish bins for ages (minimum is 7, maximum is 45)
# Categorize the existing players using the age bins. Hint: use pd.cut()
# Calculate the numbers and percentages by age group
# Create a summary data frame to hold the results
# Optional: round the percentage column to two decimal points
# Display Age Demographics Table

age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]
age_ranges = ["<10", "10-14", "15-19","20-24", "25-29", "30-34", "35-39", "40+"]

source_df = purchase_data.drop_duplicates(["SN"])
source_df["Age Demographics"] = pd.cut(source_df["Age"], age_bins, labels=age_ranges, include_lowest=True)

age_demographics_series = source_df['Age Demographics'].value_counts()

final_age_demographics_df = pd.DataFrame(age_demographics_df)

percentage_age = {}


for the_single_item in age_demographics_series.items():
    percentage_age[the_single_item[0]] = "{:.2%}".format(the_single_item[1] / total_players)
    
final_percentage_age = pd.DataFrame.from_records(percentage_age, index=[0]).transpose()


age_demographic_reference = pd.DataFrame({"Total Count": age_demographics_series, "Percentage of Players": percentage_age})

age_demographic_reference
```
